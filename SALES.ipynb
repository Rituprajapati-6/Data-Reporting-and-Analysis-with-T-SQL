<span style="background-color: transparent; font-size: 11pt; white-space: pre-wrap; color: rgb(0, 0, 0); font-family: Calibri, sans-serif;">a. Retrieve products whose list price is higher than the average unit price. </span> <span style="background-color: transparent; color: rgb(0, 0, 0); font-family: Calibri, sans-serif; font-size: 11pt; white-space: pre-wrap;">Retrieve the product ID, name, and list price for each product where the list price is higher than the average unit price for all products that have been sold.</span>

```sql
/* 7.a */
-- Selecting ProductID, Name, ListPrice where ListPrice is more than the AVG UnitPrice
SELECT P.ProductID,P.Name,P.ListPrice,SalesLT.SalesOrderDetail.UnitPrice
from 
    SalesLT.Product as P
INNER JOIN SalesLT.SalesOrderDetail
    on P.ProductID=SalesLT.SalesOrderDetail.ProductID
where 
    ListPrice > (Select AVG(SalesLT.SalesOrderDetail.UnitPrice) from SalesLT.SalesOrderDetail);
```

Retrieve the cost, list price, and average selling price for each product. </span> <span style="background-color: transparent; color: rgb(0, 0, 0); font-family: Calibri, sans-serif; font-size: 11pt; white-space: pre-wrap;">Retrieve the product ID, name, cost, and list price for each product along with the average unit price for which that product has been sold.</span>

```sql
-- /* 7.c */
-- Creating a column for the AverageSellingPrice which the average UnitPrice only
SELECT 
    P.ProductID, 
    P.Name, 
    P.StandardCost, 
    P.ListPrice,
    (select 
        AVG(UnitPrice) 
    from 
        SalesLT.SalesOrderDetail as SOD 
    where 
        SOD.ProductID = P.ProductID) 
    as AverageSellingPrice
from 
    SalesLT.Product as P
```

<span style="background-color: transparent; font-size: 11pt; white-space: pre-wrap; color: rgb(0, 0, 0); font-family: Calibri, sans-serif;">d. Retrieve products that have an average selling price that is lower than the cost. </span> <span style="background-color: transparent; color: rgb(0, 0, 0); font-family: Calibri, sans-serif; font-size: 11pt; white-space: pre-wrap;">Filter your previous query to include only products where the cost price is higher than the average selling price.</span>

```sql
-- /* 7.d */
-- Creating a column for the AverageSellingPrice which the average UnitPrice only and also displaying those greater than the avg selling price
SELECT 
    P.ProductID, 
    P.Name, 
    P.StandardCost, 
    P.ListPrice,
    (select 
        AVG(UnitPrice) 
    from 
        SalesLT.SalesOrderDetail as SOD 
    where 
        SOD.ProductID = P.ProductID) 
    as AverageSellingPrice
from 
    SalesLT.Product as P
    WHERE 
    P.StandardCost>
    (select 
        AVG(UnitPrice) 
    from 
        SalesLT.SalesOrderDetail as SOD 
    where 
        SOD.ProductID = P.ProductID);
```
